# 03 — Job Matching Model

Train a content-based filtering model using cosine similarity.

```python
import sys
sys.path.insert(0, '..')
from src.preprocessing import preprocess_seekers, save_models
from src.train_matching import generate_synthetic_data, train
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Generate or load training data
seekers, jobs = generate_synthetic_data()
print(f'Seekers: {len(seekers)}, Jobs: {len(jobs)}')
```

```python
# Preprocess seekers
seeker_features, vectorizer, scaler = preprocess_seekers(seekers)
print(f'Seeker feature dimensions: {seeker_features.shape}')
```

```python
# Compute similarity matrix
sim_matrix = cosine_similarity(seeker_features.fillna(0))
print(f'Similarity matrix: {sim_matrix.shape}')
print(f'Score range: [{sim_matrix.min():.3f}, {sim_matrix.max():.3f}]')
```

```python
# Save models
save_models(vectorizer, scaler, '../models/')
print('Models saved!')
```
