# 01 — Data Exploration

Explore the EquiPath dataset: job seekers, job postings, and applications.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('whitegrid')
```

```python
# Connect to PostgreSQL
import psycopg2
from dotenv import load_dotenv
import os

load_dotenv('../.env.example')
DATABASE_URL = os.getenv('DATABASE_URL', 'postgresql://postgres:password@localhost:5432/equipath')
```

```python
# Load job seekers
conn = psycopg2.connect(DATABASE_URL)
seekers = pd.read_sql('SELECT * FROM job_seekers', conn)
print(f'Job Seekers: {len(seekers)}')
seekers.head()
```

```python
# Load jobs
jobs = pd.read_sql('SELECT * FROM jobs WHERE is_active = true', conn)
print(f'Active Jobs: {len(jobs)}')
jobs.head()
```

```python
# Skills distribution
all_skills = [s for skills in seekers['skills'].dropna() for s in skills]
skill_counts = pd.Series(all_skills).value_counts().head(15)

plt.figure(figsize=(10, 6))
skill_counts.plot(kind='barh', color='#25a8b5')
plt.title('Top 15 Skills Among Job Seekers')
plt.xlabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Experience distribution
plt.figure(figsize=(8, 5))
seekers['experience_years'].hist(bins=10, color='#fbbf24', edgecolor='white')
plt.title('Experience Years Distribution')
plt.xlabel('Years')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
conn.close()
print('Done exploring!')
```
